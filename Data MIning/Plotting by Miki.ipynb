```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
data=load_iris()
df=pd.DataFrame(data=data.data,columns=data.feature_names)
df
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

plt.figure()
plt=scatter_matrix(df,c=data["target"],marker='o',s=10,alpha=.8,figsize=[10,10])
plt
```

```python
pd.DataFrame(data=data.target).value_counts().plot.pie()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(np.sort(df["sepal length (cm)"]), np.sort(df["sepal width (cm)"]))
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
plt.close()
```
